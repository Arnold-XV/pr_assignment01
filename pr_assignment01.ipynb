##Preprocessing: case folding, hapus tanda baca, tokenisasi, stopword removal, stemming

```python
#preprocessing: case folding, hapus tanda baca, tokenisasi, stopword removal, stemming
#note: daftar stopword ai-generated, disesuaikan dengan corpus
STOPWORDS = [
    "dan", "atau", "yang", "di", "ke", "dari", "pada", "untuk",
    "adalah", "ini", "itu", "dengan", "akan", "juga", "sudah",
    "tidak", "bisa", "ada", "dalam", "lebih", "sangat", "satu",
    "oleh", "telah", "mereka", "kita", "saya", "anda",
    "dia", "apa", "siapa", "mana", "kapan",
    "karena", "jika", "maka", "namun", "tetapi", "serta",
    "antara", "lain", "hal", "secara", "melalui", "tentang",
    "sebelum", "sesudah", "sedang", "masih", "hanya",
    "banyak", "sering", "hampir", "selalu", "pernah", "belum",
    "harus", "dapat", "bagi", "sebagai", "seperti", "hingga",
    "salah", "merupakan", "tersebut", "bahwa", "setiap",
    "semua", "agar", "tanpa", "suatu", "sehingga",
    "masing", "nya", "pun", "berbagai",
    "yg", "ga", "gak", "nggak", "ngga", "udh", "udah",
    "td", "jd", "jadi", "gitu", "gatau",
    "tapi", "banget", "bangett", "bangettt", "aja", "sih",
    "dong", "deh", "nih", "loh", "ya", "kok", "tu",
    "kalo", "kayak", "bener", "bener2", "parah",
    "so", "at", "the", "same", "time", "but",
    "buat", "bikin", "pake", "sama", "cuma", "lagi",
    "abis", "mau", "terlalu", "pas", "dapet", "kena",
    "saat", "situ", "sana", "sini", "kalian", "kami",
    "bakal", "menurutku", "menurut",
    "selain", "bahkan", "sampe", "cenderung", "relatif",
    "lumayan", "kurang",
    "terus", "begitu", "kata", "nanti", "beserta",
    "orang", "kalau", "datang", "dulu", "paling",
    "kenapa", "padahal", "tolong", "baru", "entah",
    "gimana", "saja", "sekali", "hehe", "huhu", "huhuu",
    "aku", "cuman", "kupikir", "kirain",
    "semoga", "hm", "oke", "okelah", "nah", "tau",
    "seporsi", "cobain", "sayang"
]

def preprocess(text):
    #case folding
    result = ""
    for c in text:
        if 'A' <= c <= 'Z':
            result += chr(ord(c) + 32)
        else:
            result += c
    text = result

    #hapus tanda baca, simpan huruf dan spasi saja
    result = ""
    for c in text:
        if ('a' <= c <= 'z') or c == ' ':
            result += c
        else:
            result += ' '
    text = result

    #tokenisasi
    tokens = []
    kata = ""
    for c in text:
        if c == ' ':
            if kata != "":
                tokens.append(kata)
                kata = ""
        else:
            kata += c
    if kata != "":
        tokens.append(kata)

    #hapus stopword
    filtered = []
    for t in tokens:
        if t not in STOPWORDS and len(t) > 1:
            filtered.append(t)

    #stemming secukupnya untuk mempertahankan makna kata dalam corpus
    prefixes = ["ber", "ter", "meng", "di"]
    suffixes = ["nya"]

    stemmed = []
    for w in filtered:
        r = w
        for p in prefixes:
            if r[:len(p)] == p and len(r) > len(p) + 2:
                r = r[len(p):]
                break
        for s in suffixes:
            if r[-len(s):] == s and len(r) > len(s) + 2:
                r = r[:-len(s)]
                break
        stemmed.append(r)

    return stemmed
```

##TF

```python
#hitung TF per corpus
def compute_tf(tokens):
    total = len(tokens)
    tf = {}
    for t in tokens:
        if t in tf:
            tf[t] += 1
        else:
            tf[t] = 1
    for t in tf:
        tf[t] = tf[t] / total
    return tf
```

##IDF

```python
#hitung IDF dari seluruh dokumen
def compute_idf(tokenized_docs):
    N = len(tokenized_docs)
    df = {}
    for tokens in tokenized_docs:
        seen = []
        for t in tokens:
            if t not in seen:
                seen.append(t)
                if t in df:
                    df[t] += 1
                else:
                    df[t] = 1
    idf = {}
    for term in df:
        idf[term] = math.log10(N / df[term])
    return idf
```

##TF-IDF

```python
#hitung TF-IDF
def compute_tfidf(documents):
    tokenized = [preprocess(doc) for doc in documents]
    idf = compute_idf(tokenized)

    tfidf_list = []
    for tokens in tokenized:
        tf = compute_tf(tokens)
        tfidf = {}
        for term in tf:
            tfidf[term] = tf[term] * idf[term]
        tfidf_list.append(tfidf)
    return tfidf_list, tokenized
```

```python
#preprocessing
print("Hasil preprocessing tiap dokumen:")
for i, d in enumerate(docs):
    print(f"Dok {i+1}: {preprocess(d)}")

#hitung TF-IDF
tfidf_values, tokenized_docs = compute_tfidf(docs)

print("\nTF-IDF per dokumen:")
for i, tfidf in enumerate(tfidf_values):
    print(f"\nDokumen {i+1}:")
    #urutkan berdasarkan bobot tertinggi
    sorted_terms = []
    for term in tfidf:
        sorted_terms.append((term, tfidf[term]))
    for a in range(len(sorted_terms)):
        for b in range(a+1, len(sorted_terms)):
            if sorted_terms[b][1] > sorted_terms[a][1]:
                sorted_terms[a], sorted_terms[b] = sorted_terms[b], sorted_terms[a]
    for term, score in sorted_terms:
        print(f"  {term:<20} {score:.4f}")
```
